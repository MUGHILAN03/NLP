```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
nltk.download()
```

```python
nltk.download('punkt')
```

```python
text = """The T20 World Cup is heating up! All eyes are on batting legends Virat Kohli, Rohit Sharma, and Shikhar Dhawan to provide a solid opening foundation. The middle order boasts the consistent KL Rahul and the explosive Rishabh Pant, while all-rounders Hardik Pandya and Ravindra Jadeja bring firepower with bat and ball. Jasprit Bumrah leads the bowling attack, alongside Mohammed Shami and the experienced Bhuvneshwar Kumar. While MS Dhoni might be retired, his mentorship and the guidance of veterans like Suresh Raina continue to inspire the Indian team."""
```

```python
tokens = nltk.regexp_tokenize(text, pattern="[^\W]+")
```

```python
tokens
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
stop_words.update(STOPWORDS)  # Add WordCloud's stopwords for more comprehensive removal
filtered_tokens = [word.lower() for word in tokens if word not in stop_words]
```

```python
word_counts = nltk.FreqDist(filtered_tokens)
```

```python
wordcloud = WordCloud(background_color="white", max_font_size=80)
```

```python
wordcloud.generate_from_frequencies(word_counts)
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
mask = np.array(Image.open('/content/download (1).jpeg'))

wordcloud = WordCloud(width=800, height=400, mask=mask).generate_from_frequencies(word_counts)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
